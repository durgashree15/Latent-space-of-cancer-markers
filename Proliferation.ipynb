```python
import pandas as pd
```

```python
exp1 = pd.read_csv("Data/exp1.csv")
exp1 = exp1.drop("Unnamed: 0", axis=1)
exp2 = pd.read_csv("Data/exp2.csv")
exp2 = exp2.drop("Unnamed: 0", axis=1)
```

```python
exp1 = exp1.fillna("null")
exp2 = exp2.fillna("null")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
exp1['Peptide'] = exp1['Peptide'].astype('category')
sns.kdeplot(data=exp1[exp1["Peptide"]=="N4"], x = "Proliferation", common_norm=False, label = "N4")
sns.kdeplot(data=exp1[exp1["Peptide"]=="A2"], x = "Proliferation", common_norm=False, label = "A2")
sns.kdeplot(data=exp1[exp1["Peptide"]=="Q4"], x = "Proliferation", common_norm=False, label = "Q4")
sns.kdeplot(data=exp1[exp1["Peptide"]=="T4"], x = "Proliferation", common_norm=False, label = "T4")
sns.kdeplot(data=exp1[exp1["Peptide"]=="V4"], x = "Proliferation", common_norm=False, label = "V4")
sns.kdeplot(data=exp1[exp1["Peptide"]=="G4"], x = "Proliferation", common_norm=False, label = "G4")
sns.kdeplot(data=exp1[exp1["Peptide"]=="E1"], x = "Proliferation", common_norm=False, label = "E1")
sns.kdeplot(data=exp1[exp1["Peptide"]=="null"], x = "Proliferation", common_norm=False, label = "null")
plt.title('Density Curve of Proliferation Across Target Labels')
plt.xlabel('Proliferation')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
exp1_n4 = exp1[exp1["Peptide"]=="N4"]

sns.kdeplot(data=exp1_n4[exp1_n4["Time"]==4], x = "Proliferation", common_norm=False, label = 4)
sns.kdeplot(data=exp1_n4[exp1_n4["Time"]==12], x = "Proliferation", common_norm=False, label = 12)
sns.kdeplot(data=exp1_n4[exp1_n4["Time"]==24], x = "Proliferation", common_norm=False, label = 24)
sns.kdeplot(data=exp1_n4[exp1_n4["Time"]==30], x = "Proliferation", common_norm=False, label = 30)
sns.kdeplot(data=exp1_n4[exp1_n4["Time"]==36], x = "Proliferation", common_norm=False, label = 36)
sns.kdeplot(data=exp1_n4[exp1_n4["Time"]==48], x = "Proliferation", common_norm=False, label = 48)
sns.kdeplot(data=exp1_n4[exp1_n4["Time"]==60], x = "Proliferation", common_norm=False, label = 60)
sns.kdeplot(data=exp1_n4[exp1_n4["Time"]==72], x = "Proliferation", common_norm=False, label = 72)
plt.title('Density Curve of Proliferation Across Target Labels')
plt.xlabel('Proliferation')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
average_proliferation = exp1_n4.groupby('Time')['Proliferation'].mean().reset_index()

print(average_proliferation)
```
