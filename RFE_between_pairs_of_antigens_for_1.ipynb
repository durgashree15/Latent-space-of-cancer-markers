```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score
import numpy as np
from pytorch_tabnet.tab_model import TabNetClassifier
```

**Important feature identification for separating classes 1 and 2**

```python
comp_1_2_1 = d_72_um_1[d_72_um_1["Peptide"].isin([1, 2])]
comp_1_2_2 = d_72_um_2[d_72_um_2["Peptide"].isin([1, 2])]
```

**RFE using Random Forest Classifier for classes 1 and 2**



```python
model = RandomForestClassifier()

n_features_to_select = 3
rfe = RFE(estimator=model, n_features_to_select=n_features_to_select)
rfe.fit(comp_1_2_1.drop("Peptide", axis = 1), comp_1_2_1["Peptide"])

X_train_rfe = rfe.transform(comp_1_2_1.drop("Peptide", axis = 1))
X_val_rfe = rfe.transform(comp_1_2_2.drop("Peptide", axis = 1))

model.fit(X_train_rfe, comp_1_2_1["Peptide"])

y_pred = model.predict(X_val_rfe)
accuracy = accuracy_score(comp_1_2_2["Peptide"], y_pred)
print(f'Validation Accuracy: {accuracy:.4f}')

print("Feature ranking:")
print(rfe.ranking_)

selected_features = np.where(rfe.support_)[0]
print("Selected features:")
print(selected_features)
```

**TabNetClassifier for classes 1 and 2**

```python
clf = TabNetClassifier()

clf.fit(
    comp_1_2_1.drop("Peptide", axis = 1).values,comp_1_2_1["Peptide"],
    max_epochs=30,
    patience=10,
    batch_size=1024,
    virtual_batch_size=128,
    num_workers=1,
    drop_last=False
)

y_pred = clf.predict(comp_1_2_2.drop("Peptide", axis = 1).values)

accuracy = accuracy_score(comp_1_2_2["Peptide"], y_pred)
print(f'Accuracy: {accuracy}')

import matplotlib.pyplot as plt
feature_importances = clf.feature_importances_
print("Feature Importances:")

plt.figure(figsize=(10, 5))
plt.bar(comp_1_2_2.columns[1:], feature_importances)
plt.xlabel('Feature Index')
plt.xticks(rotation=90)
plt.ylabel('Importance')
plt.title('Feature Importances from TabNet')
plt.show()

```

**Visualizing the density of features**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
dataset1 = comp_1_2_1[comp_1_2_1["Peptide"]==1].drop("Peptide", axis=1)
dataset2 = comp_1_2_1[comp_1_2_1["Peptide"]==2].drop("Peptide", axis=1)

features = dataset1.columns

warnings.filterwarnings("ignore")

fig, axes = plt.subplots(4, 6, figsize=(24, 16))
fig.suptitle('Feature Density Scatter Plots', fontsize=20)

# Iterate over features and create scatter plots
for i, feature in enumerate(features):
  row, col = divmod(i, 6)
  ax = axes[row, col]
  sns.kdeplot(dataset1[feature], ax=ax, label='E1', color='blue', shade=False)
  sns.kdeplot(dataset2[feature], ax=ax, label='G4', color='orange', shade=False)
  ax.set_title(feature)
  ax.set_xlabel('Value')
  ax.set_ylabel('Density')
  ax.legend()

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
