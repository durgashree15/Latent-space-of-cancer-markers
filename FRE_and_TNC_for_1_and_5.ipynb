```python
import numpy as np
import pandas as pd
from sklearn.utils import resample

min_class_count = d_72_um_1['Peptide'].value_counts().min()

downsampled_data = []

for label in d_72_um_1['Peptide'].unique():
    class_data = d_72_um_1[d_72_um_1['Peptide'] == label]
    downsampled_class_data = resample(class_data,
                                      replace=False,
                                      n_samples=min_class_count,
                                      random_state=42)
    downsampled_data.append(downsampled_class_data)

downsampled_data = pd.concat(downsampled_data)

print(downsampled_data['Peptide'].value_counts())
d_72_um_1 = downsampled_data
d_72_um_1
```

**Importing the required libraries**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
from sklearn.metrics import balanced_accuracy_score
import numpy as np
from pytorch_tabnet.tab_model import TabNetClassifier
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

**Important feature identification for separating classes 1 5 and 6**

```python
mmscaler = MinMaxScaler()
comp_1_5_6_1 = pd.DataFrame(mmscaler.fit_transform(d_72_um_1[d_72_um_1["Peptide"].isin([1, 5, 6])], d_72_um_1.columns))
comp_1_5_6_1.columns = d_72_um_1.columns
comp_1_5_6_1 = comp_1_5_6_1.drop("Peptide", axis=1)
comp_1_5_6_1["Peptide"] = d_72_um_1[d_72_um_1["Peptide"].isin([1, 5, 6])]["Peptide"].reset_index(drop=True)
mmscaler = MinMaxScaler()
comp_1_5_6_2 = pd.DataFrame(mmscaler.fit_transform(d_72_um_2[d_72_um_2["Peptide"].isin([1, 5, 6])], d_72_um_2.columns))
comp_1_5_6_2.columns = d_72_um_2.columns
comp_1_5_6_2 = comp_1_5_6_2.drop("Peptide", axis=1)
comp_1_5_6_2["Peptide"] = d_72_um_2[d_72_um_2["Peptide"].isin([1, 5, 6])]["Peptide"].reset_index(drop=True)
```

**RFE using Random Forest Classifier for classes 1 5 and 6**



```python
model = RandomForestClassifier()

n_features_to_select = 3
rfe = RFE(estimator=model, n_features_to_select=n_features_to_select)
rfe.fit(comp_1_5_6_1.drop("Peptide", axis = 1), comp_1_5_6_1["Peptide"].astype(int))

X_train_rfe = rfe.transform(comp_1_5_6_1.drop("Peptide", axis = 1))
X_val_rfe = rfe.transform(comp_1_5_6_2.drop("Peptide", axis = 1))

model.fit(X_train_rfe, comp_1_5_6_1["Peptide"].astype(int))

y_pred = model.predict(X_val_rfe)
accuracy = balanced_accuracy_score(comp_1_5_6_2["Peptide"].astype(int), y_pred)
print(f'Validation Accuracy: {accuracy:.4f}')

print("Feature ranking:")
print(rfe.ranking_)

selected_features = np.where(rfe.support_)[0]
print("Selected features:")
print(selected_features)

for i in selected_features:
  print(comp_1_5_6_1.columns[:-1][i])

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(comp_1_5_6_2["Peptide"].astype(int), y_pred)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['True E1', 'True T4', 'True V4'], yticklabels=['Pred E1', 'Pred T4', 'Pred V4'])
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Confusion Matrix')
plt.show()
```

**TabNetClassifier for classes 1 5 and 6**

```python
clf = TabNetClassifier()

clf.fit(
    comp_1_5_6_1.drop("Peptide", axis = 1).values,comp_1_5_6_1["Peptide"].astype(int),
    max_epochs=30,
    patience=10,
    batch_size=1024,
    virtual_batch_size=128,
    num_workers=1,
    drop_last=False
)

y_pred = clf.predict(comp_1_5_6_2.drop("Peptide", axis = 1).values)

accuracy = balanced_accuracy_score(comp_1_5_6_2["Peptide"].astype(int), y_pred)
print(f'Accuracy: {accuracy}')

import matplotlib.pyplot as plt
feature_importances = clf.feature_importances_
print("Feature Importances:")

plt.figure(figsize=(10, 5))
plt.bar(comp_1_5_6_2.columns[:-1], feature_importances)
plt.xlabel('Feature Index')
plt.xticks(rotation=90)
plt.ylabel('Importance')
plt.title('Feature Importances from TabNet')
plt.show()

```

**Visualizing the density of features**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
dataset1 = comp_1_5_6_2[comp_1_5_6_2["Peptide"]==1].drop("Peptide", axis=1)
dataset2 = comp_1_5_6_2[comp_1_5_6_2["Peptide"]==5].drop("Peptide", axis=1)
dataset3 = comp_1_5_6_2[comp_1_5_6_2["Peptide"]==6].drop("Peptide", axis=1)

features = dataset1.columns

warnings.filterwarnings("ignore")

fig, axes = plt.subplots(4, 6, figsize=(24, 16))
fig.suptitle('Feature Density Scatter Plots', fontsize=20)

# Iterate over features and create scatter plots
for i, feature in enumerate(features):
  row, col = divmod(i, 6)
  ax = axes[row, col]
  sns.kdeplot(dataset1[feature], ax=ax, label='E1', color='blue', shade=False)
  sns.kdeplot(dataset2[feature], ax=ax, label='T4', color='orange', shade=False)
  sns.kdeplot(dataset3[feature], ax=ax, label='V4', color='green', shade=False)
  ax.set_title(feature)
  ax.set_xlabel('Value')
  ax.set_ylabel('Density')
  ax.legend()

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
